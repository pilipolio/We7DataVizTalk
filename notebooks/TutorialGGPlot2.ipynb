## Basic example : the bar chart

 - `displ` engine displacement, in litres
 - `hwy` highway miles per gallon

```python
%%R -w 800 -h 600
library(ggplot2)
p <- ggplot(data=mpg, aes(x=hwy))
p <- p + geom_bar(stat='bin', binwidth=2)

print(p)
```

```python
%%R -w 800 -h 600
library(ggplot2)
p <- ggplot(data=mpg, aes(x=hwy, fill=factor(cyl)))
p <- p + geom_bar(stat='bin', binwidth=2)
#p <- p + geom_density(alpha = 1)
print(p)
```

```python
%%R -w 800 -h 600
library(ggplot2)
mpg$cyl[mpg$cyl==5] <- 4
p <- ggplot(data=mpg, aes(x=hwy, fill=factor(cyl)))
p <- p + geom_density(adjust=1.3, alpha = .4)
print(p)
```

## Using _faceting_

```python
%%R -w 800 -h 600

p <- ggplot(data=mpg, aes(x=hwy, fill=factor(cyl)))
p <- p + geom_density(adjust=2, alpha = .4)

p <- p + facet_wrap(~ manufacturer)

print(p + facet_wrap(~ manufacturer))
```

## A biplot using the iris dataset

```python
%%R -w 800 -h 600
p <- ggplot(iris, aes(x=Sepal.Length, y=Petal.Length)) + geom_point(aes(color=Species, size=Petal.Width), alpha = I(0.7))
print(p)
```

```python
%%R -w 900 -h 400
p <- ggplot(iris, aes(x=Sepal.Length, y=Petal.Length, color=Species)) 
p <- p + geom_point()
p <- p + stat_smooth(method = "lm", formula = y ~ poly(x, 2))# + facet_grid(. ~ Species) 
print(p)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(copy=True)

scaled_features = scaler.fit_transform(iris_df[iris.feature_names])
scaled_df = pd.DataFrame(data=scaled_features, columns=[f[:-5] for f in iris.feature_names])
iris_df = pd.concat((iris_df, scaled_df), axis=1)
```
