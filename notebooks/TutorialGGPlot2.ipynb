 - `ggplot2` examples based on this [basic tutorial](http://blog.echen.me/2012/01/17/quick-introduction-to-ggplot2/)

It takes care of many of the fiddly details that make plotting a hassle (like drawing legends) as well as providing a powerful model of graphics that makes it easy to produce complex multi-layered graphics."

 - Hadley Wickham [talk about `ggplot2`](http://www.r-statistics.com/tag/ggplot2/)
 - `ggplot2` [main page](http://ggplot2.org/)

## Basic example : the bar chart

 - `displ` engine displacement, in litres
 - `hwy` highway miles per gallon

```python
%%R
library(ggplot2)
p <- ggplot(data=mpg, aes(x=hwy))
p <- p + geom_bar(stat='bin', binwidth=2)

print(p)
```

```python
%%R
library(ggplot2)
p <- ggplot(data=mpg, aes(x=hwy, fill=factor(cyl)))
p <- p + geom_bar(stat='bin', binwidth=2)
#p <- p + scale_colour_discrete(cyl, low = "green", high = "red")
print(p)
```

```python
%%R

p <- ggplot(data=mpg, aes(x=hwy))
p <- p + geom_bar(stat='bin', binwidth=2)
p <- p + facet_wrap(~ manufacturer)
print(p + facet_wrap(~ manufacturer))

#geom_density(alpha = 0.2)
```

```python
%%R -i x
library(ggplot2)

p <- ggplot(data=mpg, aes(x=displ, y=hwy))
p <- p + geom_point()
print(p)
```

## Possible improvement using `Pandas` and `Sklearn`

```python
import pandas as pd
#import warnings
#warnings.filterwarnings("ignore", category=DeprecationWarning)

iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['Species'] = iris.target_names[iris.target]
iris_df['Species_token'] = iris.target
iris_df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(copy=True)

scaled_features = scaler.fit_transform(iris_df[iris.feature_names])
scaled_df = pd.DataFrame(data=scaled_features, columns=[f[:-5] for f in iris.feature_names])
iris_df = pd.concat((iris_df, scaled_df), axis=1)
```

## Using _faceting_ to dissociate `species` and modelize

 - see [here](http://sharpstatistics.co.uk/r/ggplot2-guide/)
 - and here 

```python
%%R -w 900 -h 400
p <- ggplot(iris, aes(Sepal.Length, Sepal.Width)) 
p <- p + geom_point()
p <- p + facet_grid(. ~ Species) + stat_smooth(method = "lm")
print(p)
```
